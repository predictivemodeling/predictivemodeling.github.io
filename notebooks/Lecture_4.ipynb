###Lets learn how to make different graphs and visualize the data using seaborn library

###Iris dataset and working with it

```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```
# Loading built-in Dataset iris
iris = sns.load_dataset("iris")

#Lets print the first five rows
print(iris.head())
```

```
#Lets understand the dataset.. 
iris.describe()
```

```
#box plot shows us the range of the attribute's values and their distributions
#Can also see how their values are relative to each other
#Box shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution

sns.boxplot(data=iris)
```

```
#scatter plot of variables to see how they are relative
sns.relplot(x="sepal_length", y="sepal_width", hue="species", data=iris);
```

```
#draw a scatterplot of two variables x and y; then fit the regression model
sns.regplot(x="sepal_length", y="petal_length", data=iris);
```

```
#Lets make a swarm plot to better understand the distribution of values
sns.set()
sns.swarmplot(x='species', y='petal_length', data=iris)
```

```
#Create a pairplot
sns.pairplot(iris, hue='species')
```

```
#Plot univariate or bivariate distributions using kernel density estimation.

#A kernel density estimate (KDE) plot is a method for visualizing the 
#distribution of observations in a dataset, analagous to a histogram. 
#KDE represents the data using a continuous probability density curve 
#in one or more dimensions.

#It estimates the distribution of a variable

sns.pairplot(iris, hue="species", kind="kde")
```

```
#I want to create boxplots for all the 
fig, ax = plt.subplots(2, 2, figsize=(15, 10))
vals=(list(iris.columns))
vals.remove('species')
for var, subplot in zip(vals, ax.flatten()):
    sns.boxplot(x=var, y='species', data=iris, ax=subplot)
```

```
newiris=iris.drop(columns=['species'])
ax = sns.heatmap(newiris.corr(), cmap="YlGnBu")
```
